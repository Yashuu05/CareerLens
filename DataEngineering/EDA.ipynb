```python
score_features = []
for cols in df.columns:
    if df[cols].dtypes == "float":
        score_features.append(cols)
print("float features = \n", score_features)
```

```python
def plot_histogram(feature, df) -> None:
    import matplotlib.pyplot as plt
    import seaborn as sns
    plt.figure(figsize=(7,5))
    plt.title(f"{feature} Distribution")
    sns.histplot(data=df, x=feature, kde=True)
    plt.tight_layout()
    plt.plot()

for cols in score_features:
    plot_histogram(feature=cols, df=df)
```

```python
def find_outliers(feature: str, data: any) -> int:

    
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define outlier boundaries
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Count rows that fall outside the boundaries
    outliers = data[(data[feature] < lower_bound) | (data[feature] > upper_bound)]
    
    return len(outliers)
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
print("numeric columns: \n", numeric_cols)
```

```python
numeric_cols = numeric_cols.rename("placement_status")
total_outliers = 0
for cols in numeric_cols:
    outliers = find_outliers(feature=cols, data=df)
    total_outliers = total_outliers + outliers
    print(f"{cols}: {outliers}")

print(f"total outliers in numeric columns: {int(total_outliers)}")

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_box_plot(feature, data):
    sns.set_style("whitegrid")
    plt.figure(figsize=(8,3))
    plt.title(f"{feature} box plot")
    plt.xlabel(f"<--- {feature} --->")
    sns.boxplot(data=data, x=data['cgpa'], fill=True, gap=0.1, linewidth=0.75)
    plt.plot()

for cols in numeric_cols:
    plot_box_plot(feature=cols, data=df)
```
